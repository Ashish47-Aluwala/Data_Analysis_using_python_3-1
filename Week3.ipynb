# Task
Analyze the "emails.csv" dataset to investigate the relationship between the frequency of the words "account" and "login" and whether an email is spam. Create a scatter plot showing the count of "account" on the x-axis and the count of "login" on the y-axis. Represent spam emails as red stars and non-spam emails as blue circles. Title the plot "Account & Login Word Frequency in Spam vs Non-Spam Emails" and include a legend and axis labels.

## Load the data

### Subtask:
Load the `emails.csv` dataset into a pandas DataFrame.


**Reasoning**:
Import the pandas library, load the emails.csv file into a pandas DataFrame, and display the first few rows and information to understand the data.



```python
import pandas as pd

df = pd.read_csv('/content/emails.csv')
display(df.head())
display(df.info())
```

## Feature engineering

### Subtask:
Create new columns in the DataFrame to count the occurrences of the words "account" and "login" in the email content.


**Reasoning**:
Create 'account_count' and 'login_count' columns and populate them with the values from the 'account' and 'login' columns if they exist.



```python
df['account_count'] = 0
df['login_count'] = 0

if 'account' in df.columns:
    df['account_count'] = df['account']

if 'login' in df.columns:
    df['login_count'] = df['login']

display(df[['account_count', 'login_count']].head())
```

## Data preparation

### Subtask:
Prepare the data for plotting by selecting the relevant columns (word counts and spam/non-spam indicator).


**Reasoning**:
Select the 'account_count', 'login_count', and 'Prediction' columns from the DataFrame for plotting.



```python
plot_df = df[['account_count', 'login_count', 'Prediction']].copy()
display(plot_df.head())
```

## Create scatter plot

### Subtask:
Generate a scatter plot with "account" word count on the x-axis and "login" word count on the y-axis. Use different colors and markers to distinguish between spam and non-spam emails.


**Reasoning**:
Generate a scatter plot to visualize the relationship between 'account_count', 'login_count', and 'Prediction' using different colors and markers for spam and non-spam emails, and add labels, a title, and a legend.



```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

spam_emails = plot_df[plot_df['Prediction'] == 1.0]
non_spam_emails = plot_df[plot_df['Prediction'] == 0.0]

plt.scatter(non_spam_emails['account_count'], non_spam_emails['login_count'], color='blue', marker='o', label='Non-Spam')
plt.scatter(spam_emails['account_count'], spam_emails['login_count'], color='red', marker='*', label='Spam')

plt.xlabel('Account Word Count')
plt.ylabel('Login Word Count')
plt.title('Account & Login Word Frequency in Spam vs Non-Spam Emails')
plt.legend()
plt.grid(True)
plt.show()
```

## Summary:

### Data Analysis Key Findings

*   A scatter plot was generated to visualize the relationship between the frequency of the words "account" and "login" and whether an email is spam.
*   Non-spam emails were plotted as blue circles, and spam emails were plotted as red stars.
*   The plot is titled "Account & Login Word Frequency in Spam vs Non-Spam Emails" with labeled axes and a legend.
