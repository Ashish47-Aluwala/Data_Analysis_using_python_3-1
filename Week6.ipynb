Simulate a dataset of 2000 patient systolic blood pressure readings, load it into a Pandas DataFrame, and perform the following analysis: 1. Plot a histogram of the full dataset. 2. Take random samples of different sizes (e.g., 25, 50, 100) and repeat the sampling process at least 1000 times. 3. Plot histograms of the sample means for each sample size. 4. Compute and display the mean, median, variance, and standard deviation of the full dataset. 5. Compare how the distribution of sample means narrows as the sample size increases. 6. ## Data simulation and loading

### Subtask:
Simulate 2,000 patient systolic blood pressure readings and load them into a Pandas DataFrame.


**Reasoning**:
The subtask requires simulating data and loading it into a pandas DataFrame. This can be done by importing numpy and pandas, setting a random seed, generating random numbers, and creating a DataFrame.



```python
import numpy as np
import pandas as pd

np.random.seed(42)
systolic_bp_readings = np.random.normal(loc=120, scale=10, size=2000)
df = pd.DataFrame({'Systolic_BP': systolic_bp_readings})
display(df.head())
display(df.shape)
```

## Data exploration and visualization

### Subtask:
Plot a histogram of the full dataset and generate multiple random samples of varying sizes.


**Reasoning**:
Generate a histogram of the 'Systolic_BP' column and then generate random samples of different sizes, calculate their means, and store them for further analysis.



```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.hist(df['Systolic_BP'], bins=30, edgecolor='black')
plt.title('Histogram of Full Dataset Systolic Blood Pressure')
plt.xlabel('Systolic Blood Pressure')
plt.ylabel('Frequency')
plt.show()

sample_sizes = [25, 50, 100]
sample_means = {size: [] for size in sample_sizes}

for size in sample_sizes:
    for _ in range(1000):
        sample = df['Systolic_BP'].sample(size, replace=True)
        sample_means[size].append(sample.mean())
```

## Analyze and visualize sample means

### Subtask:
Compute the means of the generated samples and plot histograms of these sample means for each sample size to observe the effect of sample size on the distribution of means.


**Reasoning**:
Iterate through the sample means for each sample size and plot a histogram for each, titling and labeling them appropriately.



```python
for size, means in sample_means.items():
    plt.figure(figsize=(8, 6))
    plt.hist(means, bins=30, edgecolor='black')
    plt.title(f'Distribution of Sample Means (Sample Size = {size})')
    plt.xlabel('Sample Mean')
    plt.ylabel('Frequency')
    plt.show()
```

## Statistical summary

### Subtask:
Calculate and display key statistical measures (mean, median, variance, standard deviation) for the full dataset and discuss the relationship between sample size and the accuracy of estimating the population mean based on the sample means.


**Reasoning**:
Calculate and print the mean, median, variance, and standard deviation of the full dataset.



```python
mean_bp = df['Systolic_BP'].mean()
median_bp = df['Systolic_BP'].median()
variance_bp = df['Systolic_BP'].var()
std_dev_bp = df['Systolic_BP'].std()

print(f'Mean of Systolic Blood Pressure: {mean_bp:.2f}')
print(f'Median of Systolic Blood Pressure: {median_bp:.2f}')
print(f'Variance of Systolic Blood Pressure: {variance_bp:.2f}')
print(f'Standard Deviation of Systolic Blood Pressure: {std_dev_bp:.2f}')

print("\nDiscussion on Sample Size and Accuracy:")
print("The histograms of sample means for different sample sizes (25, 50, 100) visually demonstrate the Central Limit Theorem.")
print("As the sample size increases, the distribution of sample means becomes narrower and more bell-shaped, centered around the true population mean.")
print("This narrowing indicates that the variability (variance and standard deviation) of the sample means decreases with larger sample sizes.")
print("The standard deviation of the sample means is also known as the standard error of the mean.")
print("A smaller standard error implies that the sample mean is a more precise estimate of the population mean.")
print("Therefore, increasing the sample size leads to a more accurate estimation of the true population mean.")
```

## Adaptation and visualization

### Subtask:
Allow the user to specify custom sample sizes and create a line plot illustrating how the standard error decreases as the sample size increases.


**Reasoning**:
Define custom sample sizes, simulate the sampling process for these sizes, calculate the standard error for each custom sample size, and then create a line plot to show how standard error decreases as sample size increases.



```python
custom_sample_sizes = [40, 75, 150]
custom_sample_means = {size: [] for size in custom_sample_sizes}

for size in custom_sample_sizes:
    for _ in range(1000):
        sample = df['Systolic_BP'].sample(size, replace=True)
        custom_sample_means[size].append(sample.mean())

standard_errors = [np.std(custom_sample_means[size]) for size in custom_sample_sizes]

plt.figure(figsize=(10, 6))
plt.plot(custom_sample_sizes, standard_errors, marker='o')
plt.title('Standard Error vs. Sample Size (Custom Sizes)')
plt.xlabel('Sample Size')
plt.ylabel('Standard Error')
plt.grid(True)
plt.show()
```

*   Histograms of sample means for sample sizes 25, 50, and 100 show that as sample size increases, the distribution of sample means becomes narrower and more centered around the true population mean.
*   A line plot using custom sample sizes (40, 75, 150) visually demonstrates that the standard error decreases as the sample size increases.
